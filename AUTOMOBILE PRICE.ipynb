```python

# price NUM ,'peak-rpm' NUM,
#'horsepower' NUM,'stroke' NUM , 'bore' NUM , 'num-of-doors' CHAR , 'normalized-losses' NUM
```

```python
def clean_categorical(column):
    mode = df[df[column]!="?"][column].astype(str).mode()
    df[column] = df[column].str.replace("?", str(mode)).astype(str)
    return df
```

```python
def clean_numerical(column):
    mean = df[df[column]!="?"][column].astype(float).mean()
    df[column] = df[column].str.replace("?", str(mean)).astype(float)
    return df
```

```python
df.head()
```

```python
X= df.drop(['make','price'],1)
```

```python
X=pd.get_dummies(X)
```

```python
X.head()
```

```python
y= df['price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# LINEAR REG

```python
#lINEAR REGRESSION
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
y_pred= lin_reg.predict(X_test)
```

```python
y_pred
```

```python
lin_reg.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(lin_reg,X_train,y_train,cv=5)
print('Average Cross Validation Score for Linear Regression : ',cross_val.mean())
```

# KNN

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsRegressor
classifier = KNeighborsRegressor(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(classifier,X_train,y_train,cv=5)
print('Average Cross Validation Score for Linear Regression : ',cross_val.mean())
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rforest = RandomForestRegressor()
rforest.fit(X_train, y_train)
pred3 = rforest.predict(X_test)
```

# LASSO

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
alphas = np.logspace(-5,5,num=15)
grid = GridSearchCV(estimator=Lasso(), param_grid=dict(alpha=alphas), cv = 5)
grid.fit(X_train, y_train)
```

```python
lasso = Lasso(alpha = grid.best_estimator_.alpha, normalize = True) # applied the best estimator
lasso.fit(X_train, y_train)
pred1 = lasso.predict(X_test)
```

```python
lasso.score(X_test , y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(lasso,X_train,y_train,cv=5)
print('Average Cross Validation Score for Linear Regression : ',cross_val.mean())
```

# RIDGE

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
alphas = np.logspace(-5,5,num=15) # range for alpha
grid2 = GridSearchCV(estimator=Ridge(), param_grid = dict(alpha=alphas), cv = 5)
grid2.fit(X_train, y_train)
```

```python
ridge = Ridge(alpha = grid2.best_estimator_.alpha, normalize = True) # applied the best estimator
ridge.fit(X_train, y_train)
pred1 = ridge.predict(X_test)
ridge.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(ridge,X_train,y_train,cv=5)
print('Average Cross Validation Score for Linear Regression : ',cross_val.mean())
```
